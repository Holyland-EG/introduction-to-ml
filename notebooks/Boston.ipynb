```python
from sklearn.datasets import load_boston
import numpy as np
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
from sklearn.preprocessing import scale
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
```

```python
data = load_boston()
```

```python
data['data']=scale(data['data'])
```

```python
steps = np.linspace(1,10,num=200)
```

```python
score = []
kfold = KFold(n_splits=5,shuffle=True,random_state=42)
for step in steps:
    kn_regr = KNeighborsRegressor(p=step, n_neighbors=5, weights='distance')
    #train model with cv of 5 
    cv_scores = cross_val_score(kn_regr, data['data'], data['target'], cv=kfold, scoring='neg_mean_squared_error')
    score.append(round(np.mean(cv_scores),2))
```
